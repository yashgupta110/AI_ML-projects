```python
data = pd.read_excel("E:\project\AI_ML\Dataset_master.xlsx",sheet_name = "Big Basket.com Cart Apriori (Pr" )
data
```

```python
transactions = []
for i in range(0,7218):
    transactions.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
transactions
```

```python
from apyori import apriori
basket = apriori(transactions=transactions,min_support = 0.003,min_confidence = 0.2, min_length = 1,min_lift = 3,max_length = 3)
```

```python
results = list(basket)
results
```

```python
import pandas as pd

def format_apriori_results(results):
    """
    Converts the output of the apyori.apriori function into a readable pandas DataFrame.
    
    Parameters:
        results (list): List of results from the apyori.apriori function.
        
    Returns:
        pd.DataFrame: A DataFrame with association rules, support, confidence, and lift.
    """
    # Initialize lists to store extracted data
    rules = []
    supports = []
    confidences = []
    lifts = []
    
    # Loop through results
    for result in results:
        support = result.support  # Extract support
        
        for ordered_stat in result.ordered_statistics:
            # Extract base and add items
            base_items = ', '.join(ordered_stat.items_base)
            add_items = ', '.join(ordered_stat.items_add)
            confidence = ordered_stat.confidence
            lift = ordered_stat.lift
            
            # Append to respective lists
            rules.append(f"{base_items} -> {add_items}")
            supports.append(support)
            confidences.append(confidence)
            lifts.append(lift)
    
    # Create and return the DataFrame
    return pd.DataFrame({
        'Association Rule': rules,
        'Support': supports,
        'Confidence': confidences,
        'Lift': lifts
    })


# Convert results to DataFrame
df_results = format_apriori_results(results)

# Display the DataFrame
print(df_results)
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("E:\project\AI_ML\Dataset_master.xlsx",sheet_name = "Netflix Movie Recommendation Ec" )
data
```

```python
trans = []
for i in range(0,7465):
    trans.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
trans
```

```python
from apyori import apriori
results = apriori(transactions=trans,min_support = 0.003,min_confidence = 0.2, min_length = 1,min_lift = 3,max_length = 3)
```

```python
import pandas as pd

def format_apriori_results(results):
    # Initialize lists to store extracted data
    movie1 = []
    movie2 = []
    supports = []
    # confidences = []
    # lifts = []
    
    # Loop through results
    for result in results:
        support = result.support  # Extract support
        
        for ordered_stat in result.ordered_statistics:
            # Extract base and add items
            base_items = ', '.join(ordered_stat.items_base)
            add_items = ', '.join(ordered_stat.items_add)
            # confidence = ordered_stat.confidence
            # lift = ordered_stat.lift
            
            # Append to respective lists
            movie1.append(base_items)
            movie2.append(add_items)
            supports.append(support)
            # confidences.append(confidence)
            # lifts.append(lift)
    
    # Create and return the DataFrame
    return pd.DataFrame({
        'Movie_1': movie1,
        'Movie_2': movie2,
        'Support': supports,
        # 'Confidence': confidences,
        # 'Lift': lifts
    })


# Convert results to DataFrame
df_results = format_apriori_results(results)

# Display the DataFrame
print(df_results)
```
